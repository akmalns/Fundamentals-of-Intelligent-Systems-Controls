# Tugas 1 Linear Regression

Pada tugas ini diperkenalkan library  pandas untuk menyediakan tools pengolahan data, dan library seaborn sebagai tools visualisasi data. 

Menggunakan dataset yang sudah tersedia pada library Scikitlearn, dilakukan analisis terhadap data beserta sebarannya. Sebaran data menunjukkan seluruh kemungkinan nilai-nilai (atau interval) yang ada dari suatu data, beserta tingkat kemunculannya.

Anda diminta untuk membangun model regresi yang dapat memprediksi median harga rumah berdasarkan fitur-fitur yang diberikan, jika diketahui dataset dengan sebaran data sedemikian rupa.



# Dataset Sklearn

Pada library scikit learn terdapat beberapa contoh dataset yang dapat digunakan untuk membangun model pembelajaran mesin (machine learning). Data-data ini umumnya sudah melalui tahap preprocessing sehingga bisa dapat langsung diolah. Pada tugas ini digunakan dataset Boston House Price yang tersusun dari variabel-variabel yang dapat berpengaruh terhadap nilai median harga rumah.

```python
# Library untuk pengolahan data
import numpy as np
import pandas as pd
import random as rnd

# Library untuk visualisasi data
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Import library untuk pembentukan model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 

# Import dataset yang disediakan oleh library Scikit Learn
from sklearn import datasets
```

```python
# Load contoh dasaset Boston House Price dari library scikit learn
boston = datasets.load_boston()

# Melihat deskripsi dataset menggunakan built-in function pada scikit learn
# Menampilkan deskripsi singkat mengenai fitur-fitur yang ada pada dataset
print(boston.DESCR)
```

```python
# Menyimpan nilai fitur pada dataset Boston ke dalam format dataframe pandas dengan nama asli
features = pd.DataFrame(boston.data, columns=boston.feature_names)

# Menampilkan nilai-nilai fitur dari 5 data teratas pada dataframe
features.head()
```

```python
# Menampilkan tipe data dari setiap fitur
print(features.dtypes)
```

```python
# Menyimpan nilai target pada dataset Boston ke variabel baru dalam format dataframe pandas yang memiliki nama "MEDV"
target = pd.DataFrame(boston.target, columns=["MEDV"])

# Menampilkan nilai-nilai fitur dari 5 data teratas pada dataframe
target.head()
```

```python
# Menampilkan tipe data dari target
print(target.dtypes)
```

```python
# Mendefinisikan objek regresi dalam bentuk python object
model = LinearRegression()

# Mendefinisikan variabel-variabel training menggunakan pasangan data-data fitur beserta targetnya
X = features
y = target

# Melatih model dengan data train
model.fit(X,y)
```

```python
# Menampilkan parameter model regresi yang dimensinya sesuai dengan dimensi fitur
print(model.intercept_)
print(model.coef_)

# Menghitung score model menggunakan R^2
print(model.score(X,y))

# Menggunakan model regresi untuk menghitung prediksi dari seluruh data input kemudian menampilkan 5 elemen pertama sebagai contoh
predictions = model.predict(X)
print(predictions[0:5])
```

# Pengolahan Data

Kali ini dataset dibagi menjadi data train dan test, dengan data train digunakan untuk membentuk representasi model regresi. Kemudian dilakukan pengujian menggunakan data train dan data test, perbedaan hasil kedua data menunjukan karakteristik model regresi yang dihasilkan (underfit/perfect fit/overfit).

```python
# Membagi dataset menjadi data train dan data test, dengan perbandingan 4:1
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)

# Menampilkan ukuran dataset yang telah dibagi untuk memastikan jumlah dataset berbanding 4:1
X_train.shape
y_train.shape
X_test.shape
y_test.shape
```

```python
# Melatih regressor dengan seluruh fitur beserta label/target nya
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Menghitung performa regressor untuk data train dan data test, kemudian membandingkan kedua nilai tersebut
train_score = regressor.score(X_train, y_train)
print('train score:', train_score)

test_score = regressor.score(X_test, y_test)
print('test score:', test_score)
```

```python
# Menyalin fitur-fitur ke dataframe baru supaya dataframe asli tidak berubah selama pengolahan data
combine = features.copy()
combine.head()
```

```python
# Menggabungkan fitur target ke dataframe baru sehingga dihasilkan dataframe yang lengkap
combine['MEDV'] = target
combine.head()
```

```python
# Memastikan bahwa seluruh fitur memiliki data yang lengkap karena data yang hilang jika digunakan untuk training dapat menurunkan performa model
combine.isnull().any()
```

```python
# Menggunakan fungsi describe() untuk menampilkan nilai-nilai statistik dari setiap fitur di dataframe
combine.describe()
```
